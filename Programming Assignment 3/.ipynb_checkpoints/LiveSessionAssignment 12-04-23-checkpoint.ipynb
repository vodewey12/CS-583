```python
import numpy as np
import pandas as pd
```

```python
#Set Seed for Reproducibility
np.random.seed(123)

#Number of Rows and Columns
num_rows = 625
num_cols = 50
half_cols = int(num_cols / 2)

#Generate Dataset
binomial_columns = np.random.binomial(n=1, p=.5, size = (num_rows, half_cols))
normal_columns = np.random.normal(loc = 0.5, scale=0.2, size=(num_rows, half_cols))
normal_columns = np.where(normal_columns < 0.5, 0, 1)
full_data = np.concatenate([binomial_columns, normal_columns], axis=1)

#Column Names
col_names = [f'col{i+1}' for i in range(num_cols)]

#Create Dataframe
df_example = pd.DataFrame(full_data, columns = col_names)
```

```python
df_example.head()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

#Split data into Train and Test
features = df_example.loc[:,'col1':'col49'].to_numpy()
target = df_example.loc[:,'col50'].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=100, test_size=0.2)

clf2 = MLPClassifier(activation = 'relu', solver='adam', alpha=1e-5,
                     hidden_layer_sizes=(20, 2), random_state=1, max_iter = 500)
clf2.fit(X_train, y_train)
```

```python
predictions = clf2.predict(X_test)

print(metrics.classification_report(y_test, predictions))
```
